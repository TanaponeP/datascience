```python
### threshold session ###

from polyglot.text import Text 
from __future__ import division
import pandas as pd
import os, errno

# -- Format Type -- #

# polyglot 
# [(pos,neg,neu)  polz  before  after  text]

# lexicon
# [(pos,neg,neu)  (pos,neg)  polz  before  after  text]


### -- Select polyglot & lexicon -- ###

s_type = "lexicon" 

### -- Project Name -- ###

project_name = "true" 

### ถ้ากำหนดประเภทเป็น lexicon ต้องระบุชื่อของ lexicon ที่ต้องการทำผลคะแนนด้วย

lexicon = 'officeM' # EXP true,office


def is_number(n):
    try:
        float(n)   
        
    except ValueError:
        
        return 0
    
    return n.strip()



if s_type == "polyglot" :
    
        x_before = 2 
        x_after = 3
        x_tu = 1
        file_new = 'sentiment_file/'+project_name+'/'+s_type+'_report/threshold/threshold.txt'
        file_target = "sentiment_file/"+project_name+"/"+s_type+"_report/"+s_type+"_"+project_name+"_summary.txt"
        file_write = 'sentiment_file/'+project_name+'/'+s_type+'_report/threshold/threshold.txt'
    
if s_type == "lexicon" :

        x_before = 3
        x_after = 4
        x_tu = 2
        file_new = 'sentiment_file/'+project_name+'/'+s_type+'_report/threshold/'+s_type+"_"+lexicon+'_threshold.txt'
        file_target = "sentiment_file/"+project_name+"/"+s_type+"_report/"+s_type+"_"+lexicon+"_summary.txt"
        file_write = 'sentiment_file/'+project_name+'/'+s_type+'_report/threshold/'+s_type+"_"+lexicon+'_threshold.txt'


### Check Folder and Files

ckdir = 'sentiment_file/'+project_name+'/'+s_type+'_report/threshold/'

try:
    os.makedirs(ckdir)
except OSError as e:
    if e.errno != errno.EEXIST:
        raise

### Close a new files

f= open(file_new,"w")


for z in [z * 0.05 for z in range(20)]:

    with open(file_target, "r") as ins:
            
        data = []
        i = 1
        bneu=0
        bpos=0
        bneg=0

        aneu=0
        apos=0
        aneg=0

        tneu=0
        tpos=0
        tneg=0

        Ax = [0]*9
        Tx = [0]*9


        for line in ins:
            
            thresholdneg = 0.0
            thresholdpos = 0.0
            thresholdneu = 0.0
            
            if z != 0.0:
        
                thresholdneg = -z 
                thresholdpos = z
    
            
            result = [x.strip() for x in line.split('\t')] 

#             result[1] = tu
#             result[2] = before
#             result[3] = after
 

            Ai = 0
            Ti = 0 
            
            
# --------------------------  Before ----------------------------------------------------
    
            before_sen = is_number(result[x_before].strip())


            if before_sen == '0' :

                bneu +=1
                bresult = 0

            elif before_sen == '1':

                bpos +=1
                Ai+=1
                bresult = 1

            elif before_sen == '-1':

                bneg +=1
                Ai+=2
                bresult = -1
                

# --------------------------  After  ----------------------------------------------------
            
            after_sen = is_number(result[x_after].strip())


            if after_sen == '0' :

                aneu +=1
                aresult = 0  

            elif after_sen == '1' :

                apos +=1
                Ai=3
                Ti=3
                aresult = 1 
                 
                

            elif after_sen == '-1' :

                aneg +=1
                Ai=6
                Ti=6
                aresult = -1

                    
# --------------------------  TU  ----------------------------------------------------
            
            total = float(result[x_tu].strip()) 

            if thresholdneg <= total <= thresholdpos:

                tneu +=1
                tresult = 0 
 
            elif total  > thresholdpos:

                tpos +=1
                Ti+=1
                tresult = 1   


            elif total  < thresholdneg:

                tneg +=1
                Ti+=2
                tresult = -1            
             
            Ax[Ai]+=1
            Tx[Ti]+=1
            i +=1 
        
        with open(file_write, 'a') as the_file:

                the_file.write("[ threshold = "+("%.2f" % z)+" ]"+"\n"+str(Ax)+"\n"+str(Tx)+"\n")  
 
        print ("[ threshold = "+("%.2f" % z)+" ]")
        print (Ax)
        print (Tx)
        
### ผลลัพธ์จะถูกเก็บไว้ที่  /sentiment_file/project_name/lexicon&polyglot/threshold/threshold.txt
```
