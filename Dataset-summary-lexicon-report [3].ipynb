```python
# Import Lib
from polyglot.text import Text 
from __future__ import division
import pandas as pd
```

```python
###-- Project Name --###

project_name = "true" # <--- Rename true,officeM ...

 ### Config Lexicon ###

lexicon = "officeM" # true,officeM ...
```

- dic_score() : ฟังก์ชั่นหาผลลัพธ์ของคำ เช่น สวัสดี = 1.24543435
- dic_word() : ฟังก์ชั่นหาผลลัพธ์ทั้งประโยค 
- dic_sentiment() : ฟังก์ชั่นหาผลรวม positive และ negative ของ Sentence เช่น (1.423234,-0.3342)
- dic_word_sum() : ฟังก์ชั่นหาผลรวมของคำ (positive+negative+neutral) ## ใช้สำหรับหาร
- dic_sentiment_sum() : ฟังก์ชั่นหาผลคะแนน (positive+negative) ## ผลคะแนนที่ได้จาก lexicon
- fraction_sum() : ฟังก์ชั่นรวมผลของ ( dic_sentiment_sum() / dic_word_sum() ) ## ผลคะแนนที่ได้จาก lexicon

```python

def dic_score(word):  

    file = open("lexicon/"+lexicon+"/lexicon.txt","r")

    for line in file:

        value = line.split("||")

        if word == value[2].strip() :
            
            try :
            
                return float(value[0])
            
            except :
                
                return 0 
            
def dic_word(sentence):
    
    blob = sentence
    
    text = Text(blob.lower())

    positive = 0
    negative =  0
    neutral = 0
    
    if sentence != '' :

        for w in text.words:

                result = dic_score(w) 

                if result is None :

                    result = 0
                
                if result < 0 : 
                    
                    negative +=1                
                
                if result == 0 :
                    
                    neutral +=1
                
                if result > 0 :
                    
                    positive +=1
         
        sentence = (positive,negative,neutral)
                
    return sentence    

def dic_sentiment(sentence):
    
    blob = sentence
    text = Text(blob.lower())

    positive = 0
    negative =  0
    neutral = 0 
    
    total = 0

    if sentence != '' :

        for w in text.words:

                result = dic_score(w) 

                if result is None :

                    result = 0
                
                if result < 0 : 
                    
                    negative = negative + float(result)                
                
                if result == 0 :
                    
                    neutral = neutral + float(result)
                
                if result > 0 :
                    
                    positive = positive + float(result)                    
                                             
        sentence = positive,negative
                
    return sentence

def dic_word_sum(sentence):
    
    blob = sentence
    text = Text(blob.lower())
    
    positive = 0
    negative =  0
    neutral = 0
    
    if sentence != '' :

        for w in text.words:

                result = dic_score(w) 

                if result is None :

                    result = 0
                
                if result < 0 : 
                    
                    negative +=1
                
                if result == 0 :
                    
                    neutral +=1
                
                if result > 0 :
                    
                    positive +=1
        
        sentence = positive+negative+neutral
                
    return sentence

def dic_sentiment_sum(sentence):
    
    blob = sentence
    text = Text(blob.lower())

    positive = 0
    negative =  0
    neutral = 0 
    total = 0

    if sentence != '' :

        for w in text.words:

                result = dic_score(w) 

                if result is None :

                    result = 0 
                
                if result < 0 : 
                    
                    negative = negative + float(result) 
                
                if result == 0 :
                    
                    neutral = neutral + float(result) 
                
                if result > 0 :
                    
                    positive = positive + float(result) 
        
        sentence = positive+negative
                
    return sentence

def fraction_sum(word):
    
    try : 
    
        total = (dic_sentiment_sum(word)/dic_word_sum(word))
    
    except:
        
        total = 0
    
    return total
    
```
