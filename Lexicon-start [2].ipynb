##  ตรวจสอบจำนวนคำซ้ำ

- Function : check_pos(),check_neg(),check_neu()

```python
def check_pos(word): 

    file = open(dir_lex+'/'+project_name+"/dic_positive.txt","r")

    for line in file:

        value = line.split(":")

        if word == value[0] :
            
            try :
            
                return value[1]
            
            except :
                
                return 0
            
def check_neg(word): 

    file = open(dir_lex+'/'+project_name+"/dic_negative.txt","r")

    for line in file:

        value = line.split(":")

        if word == value[0] :
            
            try :
            
                return value[1]
            
            except :
                
                return 0

def check_neu(word): 

    file = open(dir_lex+'/'+project_name+"/dic_neutral.txt","r")

    for line in file:

        value = line.split(":")

        if word == value[0] :
            
            try :
            
                return value[1]
            
            except :
                
                return 0
```

## หาผลลัพธ์ของคำ

 - fraction() : รวมผลลัพธ์ของคำทั้งหมด แบ่งออกเป็น positive,negative,neutral ตัวอย่าง positive=52,negative=43,neutral=89
 - lex_summary() : รวมคะแนนของแต่ละคำ
 - lex_division () : นับจำนวนคำในแต่ละด้าน เช่น สวัสดี : postive=4,negative=2,neutral=8
 - pos_summary(),neg_summary(),neu_summary() : แยกผลลัพธ์ของ lexicon ออกเป็น 3 ส่วน คือ 1.Positive,2.Neagative,3.Neutral 

```python
def fraction(params):
    
    file = open(dir_lex+"/"+project_name+"/dic_"+params+".txt","r")

    total = 0
    for line in file:

        value = line.split(":")

        try:

            total = total+int(value[1])

        except:

            total = 0
             
    return total 
```

```python
# ---------------- Term Score ---------------- #

def lex_summary(list_word):

    ppos = check_pos(list_word)
    pneg = check_neg(list_word)
    pneu = check_neu(list_word)
    
#     fix 

    fraction_pos = fraction('positive')
    fraction_neg = fraction('negative')
    fraction_neu = fraction('neutral') 
    
    if ppos is None :
        
        ppos = 0
    
    if pneg is None :
        
        pneg = 0
    
    if pneu is None :
        
        pneu = 0
 
        
    xx = (int(ppos)/fraction_pos)-(int(pneg)/fraction_neg)
    xy = (int(ppos)/fraction_pos)+(int(pneg)/fraction_neg)+(int(pneu)/fraction_neu)
    
    try:
        
        p_total = xx/xy

        return p_total
        
    except ZeroDivisionError:
            
        return 0
    
    except :
    
        return 0 


def lex_division(list_word):

    ppos = check_pos(list_word)
    pneg = check_neg(list_word)
    pneu = check_neu(list_word)
    
    if ppos is None :
        
        ppos = 0
    
    if pneg is None :
        
        pneg = 0
    
    if pneu is None :
        
        pneu = 0
        
    total = int(ppos),int(pneg),int(pneu)
        
    return total 
```

```python
def pos_summary() :

    filename = open(dir_lex+"/"+project_name+"/dic_positive.txt","r")

    for line in filename:

            value = line.split(":") 

            try:

                if lex_summary(value[0].strip()) != 0 :

                        with open(dir_lex+'/'+project_name+'/dic_positive_summary.txt', 'a') as the_file:  

                            the_file.write(str(lex_summary(value[0].strip()) )+"||"+str(lex_division(value[0].strip()) )+"||"+str(value[0])+"\n")
            except :
                
                continue          
            
def neg_summary() :

    filename = open(dir_lex+"/"+project_name+"/dic_negative.txt","r")

    for line in filename:

            value = line.split(":") 

            try:

                if lex_summary(value[0].strip()) != 0 :

                        with open(dir_lex+'/'+project_name+'/dic_negative_summary.txt', 'a') as the_file:  

                            the_file.write(str(lex_summary(value[0].strip()) )+"||"+str(lex_division(value[0].strip()) )+"||"+str(value[0])+"\n")
            except :
                
                continue
            
def neu_summary() :

    filename = open(dir_lex+"/"+project_name+"/dic_neutral.txt","r")

    for line in filename:

            value = line.split(":") 

            try:

                if lex_summary(value[0].strip()) != 0 :

                        with open(dir_lex+'/'+project_name+'/dic_neutral_summary.txt', 'a') as the_file:  

                            the_file.write(str(lex_summary(value[0].strip()) )+"||"+str(lex_division(value[0].strip()) )+"||"+str(value[0])+"\n")
            except :
                
                continue    
        
```
