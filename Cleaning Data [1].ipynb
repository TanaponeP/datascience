### Cleaning Data :  การเตรียมข้อมูล

1.ประเภทและรายละเอียดของข้อมูล 

ข้อมูลที่จะอัพโหลดต้องอยู่ในรูปแบบไฟล์ csv โดยมีรายละเอียด ดังต่อไปนี้

	1.1 ไฟล์ before.csv (Convert to utf-8)
    
         แบ่ง column ดังต่อไปนี้ 
        
         1.Sentiment
         2.Topic
         3.Description
         4.Link

	1.2 ไฟล์ after.csv (Convert to utf-8)
    
        แบ่ง column ดังต่อไปนี้
        
        1.Sentiment
        2.Topic
        3.Description
        4.Keyword
        5.Link
        

2.ตำแหน่งการอัพโหลดไฟล์ข้อมูล

    ##Root_dir = /home/ubuntu/datascience

	โดยการสร้าง directory ใหม่ที่ Root_dir/sentiment_file/[project_name]/[directory]

	** ตัวอย่างเช่น **
    
    Root_dir/sentiment_file/project_true/dir_DEC/before.csv , <--- Before file
    Root_dir/sentiment_file/project_true/dir_DEC/after.csv <--- After file
    
3.การใช้งานโปรแกรม

กำหนด Projectname , Directory และรันโปรแกรม

4.ผลลัพธ์

ผลลัพธ์จะถูกจัดเก็บไว้ที่  Root_dir/sentiment_file/[project_name]/[directory]/dataset.txt



```python
import csv

### Config Project Name ###

project_name = "officeM"
directory = "DEC"

### --------------- ###

#Target file name path : /sentiment_file/[directory]/compare/output
#Set input files name 

file_before = "before.csv"
file_after = "after.csv"

Begin_file = file_after
Follow_file= file_before

# Set parameter

Neutral = 0;
Positive = 0;
Negative = 0;
Listening = 0;
abcd = 0

with open('sentiment_file/'+project_name+'/'+directory+'/'+Begin_file) as csvfile:
    
        readCSV = csv.reader(csvfile, delimiter='|')
        i = 1; 
        for rowBeg in readCSV:  
            
                    Blink = rowBeg[4]  
                      

                    with open('sentiment_file/'+project_name+'/'+directory+'/'+Follow_file) as csvfile:
                            readCSV = csv.reader(csvfile, delimiter='|')
                            matchlInk = 0; 

                            for rowFol in readCSV:  
                                
                                    Flink = rowFol[3]
                                     

                                    if Flink == Blink :  
                                        
                                        matchlInk+=1  

                                        with open('sentiment_file/'+project_name+'/'+directory+'/'+'dataset.txt', 'a') as the_file:
            
                                            #### Begin Files
    
                                            if rowBeg[0] == 'Neutral':
                    
                                                rowBeg[0] = "0"
                                            
                                            if rowBeg[0] == 'Positive':
                    
                                                rowBeg[0] = "1"
                                            
                                            if rowBeg[0] == 'Negative':
                    
                                                rowBeg[0] = "-1"
                        
                        
                                            #### Follow Files 

                                            if rowFol[0] == 'Neutral':
                    
                                                rowFol[0] = "0"
                                            
                                            if rowFol[0] == 'Positive':
                    
                                                rowFol[0] = "1"
                                            
                                            if rowFol[0] == 'Negative':
                    
                                                rowFol[0] = "-1"
                                            
                                            if rowFol[2] :
            
                                                the_file.write("::"
                                                    +str(Blink)
                                                    +"::"+str(rowBeg[1].replace('\n', ' ').replace('\r', ' '))+" "+str(rowBeg[2].replace('\n', ' ').replace('\r', ' '))
                                                    +"::"+str(rowFol[1].replace('\n', ' ').replace('\r', ' '))+" "+str(rowFol[2].replace('\n', ' ').replace('\r', ' '))
                                                    +"::"+str(rowBeg[3].replace('\n', ' ').replace('\r', ' '))
                                                    +"::"+str(rowBeg[0].replace('\n', ' ').replace('\r', ' '))
                                                    +"::"+str(rowFol[0].replace('\n', ' ').replace('\r', ' '))
                                                    +"\n") 
                                      
                             
        print('Success')
        


```
