```python
# def dic(sentence):
from polyglot.text import Text 
from __future__ import division
import pandas as pd

directory = "officeM"

lexicon = "true"


def dic_score(word):  

    file = open("dictionary/"+lexicon+"/lexicon.txt","r")

    for line in file:

        value = line.split("||")
        
#         print value[0]

        if word == value[2].strip() :
            
            try :
            
                return float(value[0])
            
            except :
                
                return 0 
            
def dic_word(sentence):
    
    blob = sentence
    
    text = Text(blob.lower())

    positive = 0
    negative =  0
    neutral = 0
    
    if sentence != '' :

        for w in text.words:

#                 print w , w.polarity

                result = dic_score(w) 

                if result is None :

                    result = 0

#                 total = total + float(result)
                
                
#                 print w , result
                
                if result < 0 : 
                    
                    negative +=1
                    
#                     print "negative"
                
                
                if result == 0 :
                    
                    neutral +=1
                    
#                     print "neutral"
                
                if result > 0 :
                    
                    positive +=1
                    
#                     print "positive"
 
        
        sentence = (positive,negative,neutral)
                
    return sentence
    
    

def dic_sentiment(sentence):
    
    blob = sentence
    text = Text(blob.lower())

    positive = 0
    negative =  0
    neutral = 0 
    
    total = 0

    if sentence != '' :

        for w in text.words:

#                 print w , w.polarity

                result = dic_score(w) 

                if result is None :

                    result = 0

#                 total = total + float(result)
                
                
#                 print w , result
                
                if result < 0 : 
                    
                    negative = negative + float(result)
                    
#                     print "negative"
                
                
                if result == 0 :
                    
                    neutral = neutral + float(result)
                    
#                     print "neutral"
                
                if result > 0 :
                    
                    positive = positive + float(result)
                    
#                     print "positive"
                    
                
 
 
        
        sentence = positive,negative
                
    return sentence

def dic_word_sum(sentence):
    
    blob = sentence
    text = Text(blob.lower())
    
    positive = 0
    negative =  0
    neutral = 0
    
    if sentence != '' :

        for w in text.words:

#                 print (w , w.polarity)

                result = dic_score(w) 

                if result is None :

                    result = 0

#                 total = total + float(result)
                
                
#                 print w , result
                
                if result < 0 : 
                    
                    negative +=1
                    
#                     print "negative"
                
                
                if result == 0 :
                    
                    neutral +=1
                    
#                     print "neutral"
                
                if result > 0 :
                    
                    positive +=1
                    
#                     print "positive"
 
        
        sentence = positive+negative+neutral
                
    return sentence
    
    

def dic_sentiment_sum(sentence):
    
    blob = sentence
    text = Text(blob.lower())

    positive = 0
    negative =  0
    neutral = 0 
    
    total = 0

    if sentence != '' :

        for w in text.words:

#                 print w , w.polarity

                result = dic_score(w) 

                if result is None :

                    result = 0

#                 total = total + float(result)
                
                
#                 print w , result
                
                if result < 0 : 
                    
                    negative = negative + float(result)
                    
#                     print "negative"
                
                
                if result == 0 :
                    
                    neutral = neutral + float(result)
                    
#                     print "neutral"
                
                if result > 0 :
                    
                    positive = positive + float(result)
                    
#                     print "positive"
                    
                
 
 
        
        sentence = positive+negative
                
    return sentence

def fraction_sum(word):
    
    try : 
    
        total = (dic_sentiment_sum(word)/dic_word_sum(word))
    
    except:
        
        total = 0
    
    return total
    
```

```python
with open("sentiment_file/"+directory+"/"+directory+"_dataset.txt", "r") as ins:
    
    for line in ins :
        
        result = [x.strip() for x in line.split('::')] 
        
        try:
            
            word = result[3]
            
        
        except:
            
            word = 0
            
#         print word
    
#         print "["+str(dic_word(word))+"\t"+str(dic_sentiment(word))+"\t"+str(fraction_sum(word))+"\t"+result[6]+"\t"+result[5]+"\t"+word+"]"
        
        if word != '' :
            
            with open('sentiment_file/'+directory+'/'+directory+'_output-dic.txt', 'a') as the_file:  

                    the_file.write("["+str(dic_word(word))+"\t"+str(dic_sentiment(word))+"\t"+str(fraction_sum(word))+"\t"+result[6]+"\t"+result[5]+"\t"+word+"]\n")


```
